```python
# Setup
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow Function
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

#Download
#!gdown --id 1_X-V1Lp6qMAl_-9opsseieprD3Lhdq8U
```

Simple eye and face detection using haarcascade classifier

```python
face_classifier = cv2.CascadeClassifier('D:/Courses Files/Udemy Course/haarcascades/Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('D:/Courses Files/Udemy Course/haarcascades/Haarcascades/haarcascade_eye.xml')

img = cv2.imread('D:/Courses Files/Udemy Course/images/kim.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_classifier.detectMultiScale(gray, 1.3, 5)

for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (127, 0, 255), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_classifier.detectMultiScale(roi_gray, 1.5, 8)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (255,255,0), 2)

imshow('Eye and Face Detection', img)
```
